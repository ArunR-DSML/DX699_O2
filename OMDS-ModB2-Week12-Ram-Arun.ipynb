# 2. Weekly graph question

Suppose you plot the number visitors to a museum over the seven days of a week; then again the following week.  Can you suggest the advantages and disadvantages of the following two ways of presenting the data?

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
```

```python
np.random.seed(0)
time_series_0 = np.array([100, 110, 130, 120, 100, 90, 80])
time_series_1 = np.array([70, 90, 100, 100, 120, 110, 90])
fig, axs = plt.subplots(1, 2)
axs[0].plot(time_series_0)
axs[1].plot(time_series_1)
plt.subplots_adjust(wspace = 0)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

```python
plt.plot(time_series_0)
plt.plot(time_series_1)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

### <font color='plum'> ANSWER

### Method 1: Side-by-Side Plots - displays each week's visitor data on a separate, adjacent graph.

**Advantages **
* **Clarity:** Each week's visitor trend can be seen in isolation, free from the visual distraction of the other week's data. This makes it very easy to understand the pattern of visits within a single week.
* **Simplicity:** clean and straightforward, focusing on one dataset at a time.

**Disadvantages **
* **Difficult Comparison:**  challenging to make direct, precise comparisons between the two weeks. To compare visitor numbers for the same day (e.g., Wednesday), we must visually carry the data point from one plot to the other, which is inefficient and can be inaccurate.
* **Identifying Overall Trends:** Harder to spot overarching trends or shifts in visitor patterns between the two weeks without a direct overlay.

***

### Method 2: Overlaid Plots -  plots both weeks' visitor data on the same set of axes.

**Advantages **
* **Direct Comparison:** Ease of direct comparison. We can immediately see on which days one week had more visitors than the other.
* **Trend Analysis:** Good for showing how the overall weekly trends differ. For instance, you can easily see if the busiest day of the week has shifted or if attendance is generally trending up or down from one week to the next.
* **Efficiency:**  comparative view in a single chart.

**Disadvantages **
* **Visual Clutter:** If the visitor numbers for both weeks were very similar, the lines could overlap significantly, making the plot confusing and hard to read. Adding more weeks to the same plot would increase this clutter.
* **Requires a Legend**  to distinguish which line corresponds to which week. Without one, the chart is ambiguous.

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 187-205 of the Storytelling With Data book as best you can. (Chapter Eight). You do not have to get the exact data values right, just the overall look and feel.

```python

# Data for  line plots
years = [2008, 2009, 2010, 2011, 2012, 2013, 2014]
price_a = [390, 415, 420, 390, 290, 270, 260]
price_b = [355, 400, 405, 370, 280, 260, 250]

# Data for  scatter points
scatter_years = [2010, 2011, 2013]
scatter_prices = [100, 160, 98]
scatter_labels = ['C', 'D', 'E']

# --- 2. Plotting ---
# Create a figure and an axes object for  plot
fig, ax = plt.subplots(figsize=(6, 4))

# Plot the two lines for retail prices
for price in [price_a, price_b]:
    ax.plot(years, price, color = 'grey', linewidth=2.5)

# Plot the individual data points
ax.scatter(scatter_years, scatter_prices, color='#003366', s=60, zorder=5)

# --- 3. Annotations and Labels ---
ax.set_title('Retail price over time', loc='left', fontsize=12, pad=20)

# Set the labels for the x and y axes
ax.set_xlabel('Year', fontsize=10, loc='left')
ax.set_ylabel('Average price', fontsize=10, loc='top')

# Add the text annotations for the lines
ax.text(2008, 400, 'A', fontsize=10, ha='left', va='center')
ax.text(2008, 345, 'B', fontsize=10, ha='left', va='center')

# Add the text annotations for the scatter points
for i, label in enumerate(scatter_labels):
    ax.text(scatter_years[i] - 0.2, scatter_prices[i], label,
            fontsize=10, ha='right', va='center')

# Add the figure caption at the bottom
fig.text(0.125, 0.01, 'FIGURE 8.15', ha='left', fontsize=10)

# --- 4. Styling and Formatting ---
# Set the limits for the x and y axes
ax.set_xlim(2007.5, 2014.5)
ax.set_ylim(0, 500)

# Set the x-axis ticks to be the specific years
ax.set_xticks(range(2008, 2015))

# Format the y-axis tick labels to include a dollar sign
formatter = mticker.StrMethodFormatter('${x:,.0f}')
ax.yaxis.set_major_formatter(formatter)

# Remove the top and right borders (spines) of the plot
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Customize the appearance of the axis ticks
ax.tick_params(axis='both', which='major', labelsize=8)

# Adjust the layout to prevent labels from being cut off
plt.tight_layout(rect=[0, 0.05, 1, 1])
plt.show()
```
